**EDA Objective:** Identify and retain low-cost NHANES variables that may help predict costly or invasive lab biomarkers (vitamin D, HDL cholesterol, A1C).

- Data: NHANES 2013–2014 (public .xpt files)
- Targets: vitamin_d, hdl_cholesterol, a1c
- Inputs: demographics, smoking/alcohol, vitals, occupation, physical activity

```python
# === Step 1: Define all raw files to load ===
raw_files = {
    "ALQ_I": "Alcohol Use",
    "BMX_I": "Body Measures",
    "BPX_I": "Blood Pressure",
    "DEMO_I": "Demographics",
    "INQ_I": "Income",
    "OCQ_I": "Occupation",
    "PAQ_I": "Physical Activity",
    "SMQ_I": "Smoking",
    "ALB_CR_I": "Urine Albumin/Creatinine",
    "GHB_I": "Glycohemoglobin",
    "HDL_I": "HDL",
    "TRIGLY_I": "Triglycerides/LDL",
    "VID_I": "Vitamin D"
}

# === Step 2: Load all files ===
dfs = []
for code, desc in raw_files.items():
    path = f"../data/raw/{code}.xpt.txt"
    df, _ = pyreadstat.read_xport(path)
    dfs.append(df)

# === Step 3: Merge all on SEQN ===
df_merged = reduce(lambda left, right: pd.merge(left, right, on='SEQN', how='outer'), dfs)

# === Step 4: Column selection and renaming ===
target_cols_raw = ['LBXVIDMS', 'LBDHDD', 'LBXGH']
feature_cols_raw = [
    # Demographics
    'RIAGENDR', 'RIDAGEYR', 'RIDRETH1', 'INDFMPIR', 'DMDEDUC2', 'DMDCITZN', 'DMDMARTL',
    # Vitals
    'BPXSY1', 'BPXDI1', 'BMXBMI', 'BMXWAIST',
    # Smoking
    'SMQ020', 'SMQ040', 'SMQ050Q', 'SMQ080',
    # Alcohol
    'ALQ130', 'ALQ120Q',
    # Physical Activity
    'PAQ605', 'PAQ620',
    # Occupation
    'OCQ260', 'OCQ380',
    # Income
    'IND235', 'INDFMMPI', 'INQ020', 'INQ300', 'IND310'
]

all_cols = ['SEQN'] + feature_cols_raw + target_cols_raw
df_model = df_merged[all_cols].dropna(subset=target_cols_raw).copy()

# === Step 5: Rename columns ===
rename_map = {
    # Targets
    'LBXVIDMS': 'vitamin_d', 'LBDHDD': 'hdl_cholesterol', 'LBXGH': 'a1c',
    # Demographics
    'RIAGENDR': 'sex', 'RIDAGEYR': 'age', 'RIDRETH1': 'race_ethnicity', 'INDFMPIR': 'poverty_income_ratio',
    'DMDEDUC2': 'education', 'DMDCITZN': 'citizenship', 'DMDMARTL': 'marital_status',
    # Vitals
    'BPXSY1': 'systolic_bp', 'BPXDI1': 'diastolic_bp', 'BMXBMI': 'bmi', 'BMXWAIST': 'waist_circumference',
    # Smoking
    'SMQ020': 'ever_smoked_100', 'SMQ040': 'current_smoker', 'SMQ050Q': 'cigs_per_day', 'SMQ080': 'age_started_smoking',
    # Alcohol
    'ALQ120Q': 'alc_freq_past12mo', 'ALQ130': 'drinks_per_day',
    # Physical Activity
    'PAQ605': 'intense_activity_work', 'PAQ620': 'some_activity_work',
    # Occupation
    'OCQ260': 'worked_last_week', 'OCQ380': 'hours_worked_week',
    # Income
    'IND235': 'monthly_family_income', 'INDFMMPI': 'monthly_poverty_ratio',
    'INQ020': 'has_wage_income', 'INQ300': 'has_20k_savings', 'IND310': 'total_savings'
}
df_model.rename(columns=rename_map, inplace=True)

# === Step 6: Classify features ===
target_cols = ['vitamin_d', 'hdl_cholesterol', 'a1c']

initial_cat = df_model.select_dtypes(include=['object', 'category']).columns.tolist()
initial_num = df_model.select_dtypes(include=[np.number]).columns.tolist()

# Add low-cardinality numerics to categoricals
likely_cats = df_model[initial_num].nunique()
low_card_cats = likely_cats[likely_cats <= 10].index.tolist()

cat_features = list(set(initial_cat + low_card_cats))
num_features = [col for col in initial_num if col not in cat_features + target_cols]

# === Step 7: Sanity check ===
classified = set(cat_features + num_features + target_cols)
all_columns = set(df_model.columns)

missing = all_columns - classified
extra = classified - all_columns

assert len(missing) == 0, f"Unclassified columns: {missing}"
assert len(extra) == 0, f"Invalid columns: {extra}"
assert len(classified) == df_model.shape[1], "Mismatch in total classified columns."

print("✅ Data successfully loaded, merged, renamed, and classified.")
print(f"🔤 Categorical: {len(cat_features)} | 🧮 Numerical: {len(num_features)} | 🎯 Targets: {len(target_cols)}")
```

```python
# Remove invalid entries

def remove_manual_invalid_entries(df):
    """
    Removes rows with special coded values for 'refused', 'don't know', or implausible responses
    across selected features. This includes values like 777, 999, 7, or 9.
    """
    df = df.copy()

    # Define problematic values by feature
    invalid_values = {
        'drinks_per_day': [777, 999],
        'alc_freq_past12mo': [777, 999],
        'intense_activity_work': [7, 9],
        'some_activity_work': [7, 9],
        'monthly_family_income': [77, 99],
        'has_20k_savings': [7, 9],
        'total_savings': [77, 99]
    }

    # Apply filters
    for col, bad_vals in invalid_values.items():
        if col in df.columns:
            before = len(df)
            df = df[~df[col].isin(bad_vals)]
            after = len(df)
            print(f"🧹 Removed {before - after} rows from '{col}' with invalid values: {bad_vals}")

    return df


df_model = remove_manual_invalid_entries(df_model)
```

```python
# === Remove extrema based on previously defined numerical features ===

# Store original row count for comparison
original_rows = len(df_model)

# Step 1: Get per-column min and max
mins = df_model[num_features].min()
maxs = df_model[num_features].max()

# Step 2: Create masks for rows that are NOT at any min or max
not_min = ~(df_model[num_features] == mins).any(axis=1)
not_max = ~(df_model[num_features] == maxs).any(axis=1)

# Step 3: Apply combined mask
df_model = df_model[not_min & not_max].copy()

# Step 4: Report
removed_rows = original_rows - len(df_model)
print(f"✅ Removed {removed_rows} rows with min/max-clipped values from numerical features.")
```

```python
# === Use previously defined: target_vars, num_features, display_name_map ===
numeric_features = [f for f in num_features if f not in target_vars]

# === Initialize R² table ===
r2_table = pd.DataFrame(index=numeric_features, columns=target_vars, dtype=float)

# === Plot settings ===
sns.set_style("white")
n_cols = 4
n_rows = int(np.ceil(len(numeric_features) / n_cols))

for target in target_vars:
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(4 * n_cols, 3 * n_rows))
    axes = axes.flatten()

    for i, feature in enumerate(numeric_features):
        ax = axes[i]
        subset = df_model[[feature, target]].dropna()
        X = subset[[feature]].values
        y = subset[target].values

        model = LinearRegression().fit(X, y)
        y_pred = model.predict(X)
        r2 = r2_score(y, y_pred)
        r2_table.loc[feature, target] = r2

        scatter_color = '#4C72B0'        
        sns.regplot(
            x=feature,
            y=target,
            data=df_model,
            ax=ax,
            ci=95,
            scatter_kws={'alpha': 1, 's': 8, 'color': target_colors[target]},            
            line_kws={'color': 'red'}
        )

        ax.set_xlabel(display_name_map.get(feature, feature), fontsize=9)
        ax.set_ylabel(display_name_map.get(target, target), fontsize=9)
        ax.grid(False)
        ax.annotate(f"$R^2$ = {r2:.2f}", xy=(0.05, 0.90), xycoords="axes fraction", fontsize=9)

    # Remove unused axes
    for j in range(i + 1, len(axes)):
        axes[j].axis('off')

    # Shared legend
    legend_elements = [
        Line2D([0], [0], color='red', lw=2, label='Linear fit'),
        Line2D([0], [0], marker='o', color='w', markerfacecolor=target_colors[target], markersize=6, label='Data'),
        Line2D([0], [0], color='red', lw=2, alpha=0.3, label='95% CI')
    ]
    fig.legend(
        handles=legend_elements,
        loc='upper right',
        bbox_to_anchor=(0.985, 0.985),
        frameon=True,
        facecolor='white',
        edgecolor='black',
        framealpha=1,
        fontsize=8,
        handlelength=1.5,
        handletextpad=0.5,
        borderpad=0.4,
        labelspacing=0.4
    )

    plt.tight_layout()
    plt.show()

# === R² summary table with display names ===
r2_table_display = r2_table.round(3)
r2_table_display.index = [display_name_map.get(f, f) for f in r2_table_display.index]
r2_table_display.columns = [display_name_map.get(t, t) for t in r2_table_display.columns]
r2_table_display
```

```python
# === Use cleaned and defined categorical features ===
plot_cat_features = cat_features.copy()

# === Plot layout ===
n_cols = 3
n_rows = int(np.ceil(len(plot_cat_features) / n_cols))
fig, axes = plt.subplots(n_rows, n_cols, figsize=(4 * n_cols, 3.2 * n_rows))
axes = axes.flatten()

# === Base boxplot styling ===
base_box_param = dict(
    whis=(5, 95), widths=0.18, patch_artist=True,
    flierprops=dict(marker='.', markeredgecolor='black', markersize=3, linestyle='none'),
    medianprops=dict(color='black', linewidth=1),
    whiskerprops=dict(linewidth=0.8),
    capprops=dict(linewidth=0.8)
)

space = 0.25
offsets = [-space, 0, space]

for i, cat in enumerate(plot_cat_features):
    ax1 = axes[i]
    ax2 = ax1.twinx()

    df_clean = df_model[[cat] + target_vars].dropna()
    if df_clean[cat].nunique() < 2:
        ax1.axis('off')
        continue

    categories = manual_orders.get(cat, sorted(df_clean[cat].dropna().unique()))
    nb_groups = len(categories)

    # Clean category list and group data
    df_clean[cat] = pd.Categorical(df_clean[cat], categories=categories, ordered=True)

    # Group target values by category (avoids pivot bug)
    target_data = {
    t: [df_clean[df_clean[cat] == c][t].dropna() for c in categories]
    for t in target_vars
}

    # Plot: left y-axis
    for idx, t in enumerate(target_vars[:2]):
        ax1.boxplot(
            target_data[t],
            positions=np.arange(nb_groups) + offsets[idx],
            boxprops=dict(facecolor=target_colors[t]),
            **base_box_param
        )

    # Plot: right y-axis
    ax2.boxplot(
        target_data[target_vars[2]],
        positions=np.arange(nb_groups) + offsets[2],
        boxprops=dict(facecolor=target_colors[target_vars[2]]),
        **base_box_param
    )

    # Axes labels
    ax1.set_xticks(np.arange(nb_groups))
    ax1.set_xticklabels(categories, rotation=30, ha='right', fontsize=7)
    ax1.set_xlabel(display_name_map.get(cat, cat), fontsize=8)
    # Separate y-axis labels for each target (left and right)
    ax1.set_ylabel(
        f"{display_name_map[target_vars[0]]} & {display_name_map[target_vars[1]]}",
        fontsize=7
    )
    
    ax2.set_ylabel(
        display_name_map[target_vars[2]],
        fontsize=7,
        color=target_colors[target_vars[2]]
    )

    ax1.tick_params(axis='y', labelsize=6)
    ax2.tick_params(axis='y', labelsize=6)
    ax1.tick_params(axis='x', labelsize=6)

# Turn off unused axes
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

# Shared legend
handles = [
    plt.Line2D([0], [0], color=target_colors[t], lw=5, label=display_name_map[t])
    for t in target_vars
]
fig.legend(
    handles=handles,
    title='Target',
    fontsize=7,
    title_fontsize=8,
    loc='lower center',
    ncol=3,
    frameon=False
)

plt.tight_layout(rect=[0, 0.05, 1, 1])
plt.show()
```
